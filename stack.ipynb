```python
class Stack:
    def __init__(self):
        self.__data=[]
    def push(self,item):
        self.__data.append(item)

    def pop(self):
        if self.isEmpty():
            print("stack is empty")
            return
        return self.__data.pop()
    def size(self):
        return len(self.__data)

    def top(self):
        if self.isEmpty():
            print("stack is empty")
            return
        return len(self.__data)-1

    def isEmpty(self):
        return len(self.__data)==0

s=Stack()
s.push(1)
s.push(2)
s.top()
s.push(3)
print(s.pop())
print(s.pop())
print(s.pop())
print(s.pop())
print(s.isEmpty())
print(s.top())
    
```

```python
# stack using linked list

class node:
    def __init__(self,data):
        self.data=data
        self.next=None

class Stack:
    def __init__(self):
        self.__head=None
        self.__count=0
    
    def push(self,item):
        newNode=node(item)
        newNode.next=self.__head
        self.__head=newNode
        self.__count+=1

    def pop(self):
        if self.size()==0:
            print("stack is empty")
            return
        data=self.__head.data
        self.__head=self.__head.next
        self.__count-=1
        return data

    def size(self):
        return self.__count

    def top(self):
        if self.size()==0:
            print("stack is empty")
            return
        return self.__head.data

    def isEmpty(self):
        return self.size()==0
s=Stack()
s.push(1)
s.push(2)
s.top()
s.push(3)
print(s.pop())
print(s.pop())
print(s.pop())
print(s.pop())
print(s.isEmpty())
print(s.top())

```

```python
# balanced paraenthesis
# whoever comes in last closed first ({})

def isBalanced(string):
    s=[]

    for char in string:
        if char in "({[":
            s.append(char)
        elif char =="}":
            if (not s or  s[-1]!="{"):
                return False
            else:
                s.pop()
        elif char ==")":
            if (not s or  s[-1]!="("):
                return False
            else:
                s.pop()
        elif char =="]":
            if (not s or  s[-1]!="["):
                return False
            else:
                s.pop()
    if (not s):
        return True
    else:
        return False

print(isBalanced(""))

```
