```python
# n-queen problem

def isSafe(row,col,board,n):
    i=row-1
    while(i>=0):  # o(n)
        if board[i][col]==1:
            return False
        i-=1
    i=row-1
    j=col-1

    while(i>=0 and j>=0):  # o(n)
        if board[i][j]==1:
            return False
        i-=1
        j-=1

    i=row-1
    j=col

    while(i>=0 and j<n): #o(n)
        if board[i][j]==1:
            return False
        i-=1
        j+=1
    return True


def printPathHelper(row,n,board):

    if row==n:
        print(board)
        print()
        return

    for col in range(0,n):
        if isSafe(row,col,board,n):
            board[row][col]=1
        printPathHelper(row+1,n,board)
        board[row][col]=0
    
def printPaths(n):
    board=[[0 for i in range(n)] for j in range(n)]
    printPathHelper(0,n,board)

n=int(input())
printPaths(n)
```
