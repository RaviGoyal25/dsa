```python
import heapq

class BinaryTreeNode:

    def __init__(self,val,freq):
        self.value=val
        self.freq=freq
        self.left=None
        self.right=None

    def __lt__(self,other):
        return self.freq<other.freq

    def __eq__(self,other):
        return self.freq==other.freq


class HuffmanCoding:

    def __init__(self,path):
        self.path=path
        self.heap=[]
        self.codes={}
        self.reverseCodes={}

   

    def frquencyDict(self,text):
        freq_dict={}
        for char in text:
            freq_dict[char]=freq_dict.get(char,0)+1
        return freq_dict

    def buildHeap(self,freq_dict):
        for key in freq_dict:
            binaryTreeNode=BinaryTreeNode(key,freq_dict[key])
            heapq.heappush(self.heap,binaryTreeNode)

    def buildTree(self):
        while(len(self.heap)>1):
            binrayTreeNode1=heapq.heappop(self.heap)
            binrayTreeNode2=heapq.heappop(self.heap)
            freqSum=binrayTreeNode1.freq+binrayTreeNode2.freq
            newNode=BinaryTreeNode(None,freqSum)
            newNode.left=binrayTreeNode1
            newNode.right=binrayTreeNode2
            heapq.heappush(self.heap,newNode)
        return

    def buildCodesHelper(self,root,curr_bits):
        if root is None:
            return
        if root.value is not None:
            self.codes[root.value]=curr_bits
            self.reverseCodes[curr_bits]=root.value
        self.buildCodesHelper(root.left,curr_bits+"0")
        self.buildCodesHelper(root.right,curr_bits+"1")


            
    def buildCodes(self):
        root=heapq.heappop(self.heap)
        self.buildCodesHelper(root,"")

    def getEncodedText(self,text):
        encodedText=""
        for char in text:
            encodedText+=self.codes[char]
        return encodedText

    def getPadddedEncodedText(self,encodedText):
        paddedAmount=8-(len(encodedText)%8)
        for i in range(paddedAmount):
            encodedText+="0"

        paddedInfo="{0:08b}".format(paddedAmount)
        return paddedInfo+encodedText

    def getBytesArray(self,paddedEncodedText):
        array=[]

        for i in range(0,len(paddedEncodedText),8):
            binstr=paddedEncodedText[i:i+8]
            array.append(int(binstr,2))
        return array

    def removePadding(self,text):
        paddingInfo=text[0:8]
        extraPadding=int(paddingInfo,2)
        text=text[8:]
        text_after_padding_removed=text[:-1*extraPadding]
        return text_after_padding_removed

    def decodeText(self,text):
        decode_text=""
        current_bits="" 
        for bit in text:
            current_bits+=bit
            if current_bits in self.reverseCodes:
                character=self.reverseCodes[current_bits]
                decode_text+=character
                current_bits=""

        
        return decode_text



        

    def compress(self):
        # get file from path
        # read text from file
        text="kjnsdkjbkjbkjbaCKJcbk"

        # make frequency dictionary using text
        freq_dict=self.frquencyDict(text)

        # construct heap from frequency dictionary
        self.buildHeap(freq_dict)

        # construct the binary tree from heap
        self.buildTree()

        # construct codes from binary tree
        self.buildCodes()

        # creating the encoded text using codes
        encodedText=self.getEncodedText(text)

        # pad this encoded text
        paddedEncodedText=self.getPadddedEncodedText(encodedText)
        
        # bytesArray=self.getBytesArray(paddedEncodedText)

        # bytes(bytesArray)

        print(self.decodeText(self.removePadding(paddedEncodedText)))


h=HuffmanCoding("path")
h.compress()




        

```
